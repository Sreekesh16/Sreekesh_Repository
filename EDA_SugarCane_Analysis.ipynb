# Data preprocessing

Convert the columns into the appropriate data types

```python
#always use as type to convert the data types of the columns
for i in range(0,len(df["Production (Tons)"])):
    df["Production (Tons)"][i]=df["Production (Tons)"][i].replace(".","")
print(df["Production (Tons)"][0])
df["Production (Tons)"]=df["Production (Tons)"].astype(float)
```

```python
df.dtypes
```

```python
for i in range(0,len(df["Production per Person (Kg)"])):
    df["Production per Person (Kg)"][i]=df["Production per Person (Kg)"][i].replace(".","")
    df["Production per Person (Kg)"][i]=df["Production per Person (Kg)"][i].replace(",",".")
df["Production per Person (Kg)"]=df["Production per Person (Kg)"].astype(float)
```

```python
df.dtypes
```

```python
df.head()
```

```python
for j in range(0,len(df["Acreage (Hectare)"])):
    df["Acreage (Hectare)"][j]=df["Acreage (Hectare)"][j].replace(".","")
```

```python
df["Acreage (Hectare)"]=df["Acreage (Hectare)"].astype(float)
```

```python
df.dtypes
```

```python
df.head()
```

```python
for k in range(0,len(df["Yield (Kg / Hectare)"])):
    df["Yield (Kg / Hectare)"][k]=df["Yield (Kg / Hectare)"][k].replace(",",".")
```

```python
df["Yield (Kg / Hectare)"][0]
```

```python
df["Yield (Kg / Hectare)"]=df["Yield (Kg / Hectare)"].astype(float)
```

```python
df.dtypes
```

```python
df["Yield (Kg / Hectare)"]=df["Yield (Kg / Hectare)"].astype(str)
```

```python
df.dtypes
```

```python
df["Yield (Kg / Hectare)"]=df["Yield (Kg / Hectare)"].astype(float)
```

```python
df.dtypes
```

```python
df.head()
```

```python
for i in range(0,len(df["Yield (Kg / Hectare)"])):
    df["Yield (Kg / Hectare)"][i]=df["Yield (Kg / Hectare)"][i].replace(".","")
    df["Yield (Kg / Hectare)"][i]=df["Yield (Kg / Hectare)"][i].replace(",",".")
```

```python
df["Yield (Kg / Hectare)"]=df["Yield (Kg / Hectare)"].replace("83.034,2","83034.2")
```

```python
df["Yield (Kg / Hectare)"]=df["Yield (Kg / Hectare)"].astype(float)
```
